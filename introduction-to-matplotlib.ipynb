## Matplotlib Example workflow

```python
# 0.import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1,2,3,4,5]
y = [10,20,30,40,50]

# 2. setup plot
fig,ax = plt.subplots(figsize=(10,10))
 
# 3. plot data
ax.plot(x,y)

# 4. customize plot
ax.set(title = "simple plot",xlabel = "x-axis",ylabel = "y-axis")

# 5.save and show
fig.savefig("images/sample.png")
```

### two options for subplot

```python
# first method
fig , ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))
# lets plot to each axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.hist(np.random.randn(1000));
ax4.bar(momo_price.keys(),momo_price.values());
```

```python
# second method
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize = (15,10)) #width = 15 and height = 10
ax[0,0].plot(x,x+2);
ax[0,1].scatter(x,np.sin(x));
ax[1,0].hist(np.random.randn(1000));
ax[1,1].bar(momo_price.keys(),momo_price.values());
```

## Which one you should choose ? (pyplot vs matplotlib oo method)
 * when plotting something quickly , okay to use pyplot method
 * when plotting something advanced , use the Object Oriented method

```python
# lets make a dataframe whoose age is over 50
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# object oriented method mixed with pyplot
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
             c='target',
            ax=ax)
ax.set_xlim([45,90]);
```

```python
# OO method from scratch
fig,ax = plt.subplots(figsize = (10,6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                    y = over_50["chol"],
                    c= over_50["target"])
#customizing the plot
ax.set(title="Heart Disease and cholestrol levels",
      xlabel="Age",
      ylabel = "chol")
# add a legend
ax.legend(*scatter.legend_elements(),title = "Target") #find legend element from c parameter in scatter variable

# add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle = "--");
```

```python
over_50.head()
```

```python
# subplot of age,cholestrol and thalach(max heart rate)
fig, (ax0,ax1) = plt.subplots(nrows = 2,
                             ncols=1,
                             figsize = (10,10),
                             sharex=True)
# add the data to ax0
scatter = ax0.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"])
ax0.legend(*scatter.legend_elements(),title = "Target")
ax0.set(title="heart disease and cholestrol",
       
       ylabel = "chol")
ax0.axhline(y = over_50["chol"].mean(),
           linestyle = "--");
# add the data to ax1
scatter1 = ax1.scatter(x = over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"])
ax1.legend(*scatter1.legend_elements(),title = "Target")
ax1.set(title = "heart disease and max heart rate",
       xlabel = "Age",
       ylabel = "Max heart rate")
ax1.axhline(y = over_50["thalach"].mean(),
           linestyle = "--");

# add a title to fig
fig.suptitle("Heart Disease Analysis",fontsize = 16,fontweight = "bold");
```

```python
# OO method from scratch
plt.style.use('seaborn-v0_8-whitegrid')
fig,ax = plt.subplots(figsize = (10,6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                    y = over_50["chol"],
                    c= over_50["target"],
                    cmap='winter')  # this change colors scheme
#customizing the plot
ax.set(title="Heart Disease and cholestrol levels",
      xlabel="Age",
      ylabel = "chol")
# add a legend
ax.legend(*scatter.legend_elements(),title = "Target") #find legend element from c parameter in scatter variable

# add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle = "--");
```

```python
# subplot of age,cholestrol and thalach(max heart rate)
fig, (ax0,ax1) = plt.subplots(nrows = 2,
                             ncols=1,
                             figsize = (10,10),
                             sharex=True)
# add the data to ax0
scatter = ax0.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"],
                     cmap = 'winter')
ax0.legend(*scatter.legend_elements(),title = "Target")
ax0.set(title="heart disease and cholestrol",
       
       ylabel = "chol")
# changing x axis limitation
ax0.set_xlim([50,80])
ax0.axhline(y = over_50["chol"].mean(),
           linestyle = "--");
# add the data to ax1
scatter1 = ax1.scatter(x = over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"],
                      cmap = 'winter')
ax1.legend(*scatter1.legend_elements(),title = "Target")
ax1.set(title = "heart disease and max heart rate",
       xlabel = "Age",
       ylabel = "Max heart rate")
# changing ax1 y axis limit x axis limit is same as above as it shares same x axis
ax1.set_ylim([60,200])
ax1.axhline(y = over_50["thalach"].mean(),
           linestyle = "--");

# add a title to fig
fig.suptitle("Heart Disease Analysis",fontsize = 16,fontweight = "bold");
```
